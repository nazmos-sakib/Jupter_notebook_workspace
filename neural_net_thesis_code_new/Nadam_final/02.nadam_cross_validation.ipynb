```python
#Risk score

def risk__score_calculate(data):
    risk_score = 0
    
    # 0
    #age
    #15-30=1
    #30-40=2
    #40-55=3
    #above >= 55 =4
    
    if(int(data[0])>=15 and int(data[0])<30):
        risk_score = risk_score + 1
        
    elif(int(data[0])>=30 and int(data[0])<40):
        risk_score = risk_score + 2
        
    elif(int(data[0])>=40 and int(data[0])<55):
        risk_score = risk_score + 3
        
    elif(int(data[0])>=55):
        risk_score = risk_score + 4
    
    #print(risk_score)
    
    # 1
    #gender
    #male-1=4
    #female-2=2
    
    if(int(data[1])==1):
        risk_score = risk_score + 4
        
    elif(int(data[1])==2):
        risk_score = risk_score + 2
    
    #print(risk_score)
          
    # 2
    #smoking
    #yes" : 1  =4
    #"no" : 0  =1
    #"ex" : 2  =3
    
    if(int(data[2])==1):
        risk_score = risk_score + 4
        
    elif(int(data[2])==0):
        risk_score = risk_score + 1
    
    elif(int(data[2])==2):
        risk_score = risk_score + 3
        
    #print(risk_score)
    
    # 3
    #'HTN', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[3])==1):
        risk_score = risk_score + 4
        
    elif(int(data[3])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 4
    #'DLP', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[4])==1):
        risk_score = risk_score + 4
        
    elif(int(data[4])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 5
    #'DM', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[5])==1):
        risk_score = risk_score + 4
        
    elif(int(data[5])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 6
    #'Physical Exercise',
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[6])==1):
        risk_score = risk_score + 4
        
    elif(int(data[6])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 7
    #'Family History', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[7])==1):
        risk_score = risk_score + 4
        
    elif(int(data[7])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 8
    #'Drug History', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[8])==1):
        risk_score = risk_score + 4
        
    elif(int(data[8])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 9
    #'Psychological Stress', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[9])==1):
        risk_score = risk_score + 4
        
    elif(int(data[9])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 10
    #'Chest Pain',
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[10])==1):
        risk_score = risk_score + 4
        
    elif(int(data[10])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 11
    #'Dyspnea', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[11])==1):
        risk_score = risk_score + 4
        
    elif(int(data[11])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 12
    #'Palpitation', 
    #"yes" : 1, =4
    #"no" : 0, =2
    
    if(int(data[12])==1):
        risk_score = risk_score + 4
        
    elif(int(data[12])==0):
        risk_score = risk_score + 2
    
    #print(risk_score)
    
    # 13
    #'ECG',
    #"abnormal" : 1, =4
    #"normal" : 0, =2
    
    if(int(data[13])==1):
        risk_score = risk_score + 4
        
    elif(int(data[13])==0):
        risk_score = risk_score + 2
    
    return risk_score
    
    
    
```
