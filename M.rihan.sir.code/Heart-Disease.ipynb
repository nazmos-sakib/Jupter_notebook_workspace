```python
total_list =[]

header_list = []
with open("data-835.csv","r") as heart_disease_data:
#     header_list = next(heart_disease_data)
    for line in heart_disease_data:
        data_list = line.split(',')
        temp_list =[]
        for k in range(len(data_list)):
            single_val = data_list[k]
            single_val = single_val.rstrip("\n")
            temp_list.append(single_val)
        total_list.append(temp_list)
#         single_row_list.append(data_list)
#         total_list.append()
print(len(total_list))
# print(header_list)

for i in range(40):
    print(total_list[i])
```

```python
# split into train and test sets
train_size = int(len(dataset) * 0.70)
test_size = len(dataset) - train_size   
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

print(len(train), len(test))
```

```python
X_train = train[:,0:15]
Y_train = train[:,17:20]

X_test = test[:,0:15]
Y_test = test[:,17:20]
```

```python
num_epoch = 1000
```

```python
# Create Model 

model = Sequential() 
model.add(Dense(32, input_shape=(15,) , activation= 'relu'))
model.add(Dense(16, activation= 'relu'))
model.add(Dense(3 , activation = 'softmax'))

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=False)

model.compile(loss='categorical_crossentropy',
                  optimizer=sgd,
                  metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, Y_train, nb_epoch=num_epoch)
```
